create a program that will allow you to enter events organizable by hour. There must be menu options of some form, and you must be able to easily edit, add, and delete events without directly changing the source code.

(note that by menu I dont necessarily mean GUI. as long as you can easily access the different options and receive prompts and instructions telling you how to use the program, it will probably be fine)

```python
from enum import Enum

# Assumptions
# Here to simplify the problem we will assume that an event fit in one hour
# and that we can only book an event for the given day
# that two events can be booked in the same day

# Variable Initialization
class Commands(Enum):
    '''
        Used to keep track of the commands available to the user
    '''
    CREATE = 0
    EDIT = 1
    DELETE = 2
    STOP = 3
    SHOW = 4
    
class Event():
    
    def __init__(self, time, content):
        self.time = time
        self.content = content
        
    def __str__(self):
        return f"Time: {self.time} \nDescription: {self.content}\n"
    
# A dictionary will hold the events with a key -> value pair
events = {}
command = ""

# We wil iterate and repeatedly ask for inputs for the user
# if the user enter one of the valid commands we will do the commands
# otherwise we simply output a helper print statement for the commands
while(True):
    
    command = input("Enter a command: ").upper()
    
    if command == Commands.CREATE.name:
        key = input("Enter a name for event: ")
        time = input("At what hour will the event take place? ")
        content = input("What is the event about? ")
        events[key] = Event(time, content)
        
    elif command == Commands.EDIT.name:
        key = input("Which event do you want to modify? ")
        
        if key in events:
            print(f"Current content is this: {events[key]}")
            content = input("What is the event about now?")
            time = input("At what hour will the event take place? ")
            events[key] = Event(time, content)
        else:
            print(f"{key} was never added to the events, create it first.")
            
    elif command == Commands.DELETE.name:
        key = input("Which event do you want to delete? ")
        events.pop(key, None)
        
    elif command == Commands.SHOW.name:
        
        # Print all the events with their key and content
        for key,content in events.items():
            print(f"{key} :\n{content}")
            print("---")
    elif command == Commands.STOP.name:
        print("See ya!")
        break
    else:
        print("The commands are: \ncreate, edit, delete, help, stop!")
```
